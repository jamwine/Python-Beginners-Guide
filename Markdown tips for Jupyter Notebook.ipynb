<h1 align="left"><font color="purple" size=5>Formatting</font></h1>


> Italics: `*asterisks*` or `_underscores_`  (word is enclosed in between asterisk or underscore)<br/>
>
> Bold: `**asterisks**` or `__underscores__` (word is enclosed in between two asterisks or two underscores)<br/>
>
> Bold & Italics both: `___Both___` or `***Both***` or `**_Both_**` or `__*Both*__`<br/>
>
> Strikethrough: `~~Scratched text.~~` (word is enclosed in between two tildes)<br/>

For example, consider these cases:

```
*This line is in Italics*
**This line is Bold**
***This line is Bold and in Italics***
~~This line is Strikethrough~~
```

*This line is in Italics*<br/>
**This line is Bold**<br/>
***This line is Bold and in Italics***<br/>
~~This line is Strikethrough~~ <br/>

Note that **Inline HTML tags** are also used in markdown. For example:<br/>
> `<b>Bold</b>`<br/>

<h1 align="left"><font color="purple" size=5>Blockquotes</font></h1>


Blockquotes are described using the (\>) sign. They are very handy in email to emulate reply text. For example:

```
> Text 1
> Text 2
```
> Text 1
> Text 2

Notice that the Text 1 and Text 2 are the part of the same blockquote. Moreover, the normal text can be written here (which is not a part of blockquote) by leaving a blank line after the blockquote. For different lines, a break html tag `<br/>` is used. Also, by inserting a (**\> **) sign or by leaving a blank line, we can separate lines.
```
> Text 1<br/>
> Text 2
```
> Text 1<br/>
> Text 2

```
> Text 1
>
> Text 2
```

> Text 1
>
> Text 2

Nesting of blockquotes is also possible using the following syntax:

```
> Level 1
>
> > Level 2
>
> Level 1
```

>

`3. Item 2`<br/>

`* Item 1`<br/>

<h1 align="left"><font color="purple" size=5>Links</font></h1>


Links can be placed in two different ways: **Inline-style link** or **reference-style link** (with or without title).

For **inline-style link**,

```
[Google Link](https://www.google.com)

[Google Link with title](https://www.google.com "Google.com")
```
[Google Link](https://www.google.com)

[Google Link with title](https://www.google.com "Google.com")

For **reference-style link**,

```
[Local Link 1](Some local link in our directory)

[Local Link 2](Some local link in our directory "Some Title")
```
[Local Link 1](Some local link in our directory)

[Local Link 2](Some local link in our directory "Some Title")


Further, we can use tags and numbers in order to define links. For example:

```
[Click here to search on Yahoo][Yahoo link]

[Click here to search on Google][1]

[1]: https://www.google.com
[Yahoo link]: http://search.yahoo.com/
```

<h1 align="left"><font color="purple" size=5>Mathematical Formulae</font></h1>


LaTeX notation is used: `$...$` for inline mathematics and `$$...$$` for displayed mathematics. The beginning and ending dollar signs ($) are the delimiters for the TeX markup. The spaces are trimmed inside the texts. Some examples are:<br/> 

`$Inline Example$`<br/>
$Inline Example$<br/>

However, escape sequences **\\** are used in order to print space in between the text.<br/>

`$Inline\ Example$`<br/>
$Inline\ Example$<br/>

`$$Displayed Example$$`
$$Displayed Example$$

`$$Displayed\ Example$$`
$$Displayed\ Example$$

Consider a mathematical equation:

<h1 align="left"><font color="blue" size=3>Inline Mathematics Syntax (using single \$ sign)</font></h1>

```
$-b \pm \sqrt{b^2 - 4ac} \over 2a$
$x = a_0 + \frac{1}{a_1 + \frac{1}{a_2 + \frac{1}{a_3 + a_4}}}$
$\forall x \in X, \quad \exists y \leq \epsilon$
```
<h1 align="left"><font color="green" size=3>Output</font></h1>

$-b \pm \sqrt{b^2 - 4ac} \over 2a$<br/>

$x = a_0 + \frac{1}{a_1 + \frac{1}{a_2 + \frac{1}{a_3 + a_4}}}$<br/>

$\forall x \in X, \quad \exists y \leq \epsilon$

<h1 align="left"><font color="blue" size=3>Displayed Mathematics Syntax (using two \$ signs)</font></h1>

```
$$\sqrt{x_1^5} \pm \frac{x_2^5}{x_1x_2} \over 4 $$<br/>
$$ \quad\forall x \in X,  \quad x_1 \leq 5,   \quad x_2 \geq 5 $$
```

<h1 align="left"><font color="green" size=3>Output</font></h1>

$$\sqrt{x_1^5} \pm \frac{x_2^5}{x_1x_2} \over 4 $$<br/>
$$ \quad\forall x \in X,  \quad x_1 \leq 5,   \quad x_2 \geq 5 $$
